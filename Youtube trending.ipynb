### plotting relation of count with increasing date

```python
a=mydata.groupby(['Upload Date'])['country'].count()

b=[0 for x in range(len(a))]
for i in range(len(a)):
    b[i]=a[i]
    
date=mydata.sort_values('Upload Date')['Upload Date'].unique()

plt.figure(figsize=(15,6),dpi=400)
#plt.style.use('ggplot')

plt.plot(date,b,label='count',linewidth=2,marker='o',markersize=10,markeredgecolor='gray')
plt.title('no of trending videos per day ',fontdict={'fontsize':20,'fontname':'Times New Roman','color':'black'})
plt.ylabel('count',fontdict={'fontsize':20,'fontname':'Times New Roman','color':'black'})
plt.xlabel('date',fontdict={'fontsize':20,'fontname':'Times New Roman','color':'black'})

plt.legend()
plt.show()
```
